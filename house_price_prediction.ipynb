### DataFrameの各カラムどうしの相関係数を比較する

* 数値データの38カラムの総当たりで相関係数を算出
* 相関係数のイメージ([wiki](https://ja.wikipedia.org/wiki/%E7%9B%B8%E9%96%A2%E4%BF%82%E6%95%B0)参照)

![相関係数](https://upload.wikimedia.org/wikipedia/commons/d/d4/Correlation_examples2.svg)

```python
corrmat = df.corr()
corrmat
```

```python
f, ax = plt.subplots(figsize=(12,9))
sns.heatmap(corrmat, vmax=.8, square=True)
```

* 対角線上の色の濃いところはidどうし、MSSubClassどうし、など同じカラムの相関係数なので1となっている。
* 対角線をまたいで反対側は、項目を入れ替えて相関を見ているだけなので、線対称になっている。
* 色が濃いところが相関ありそう。GarageYrBlt・GarageCars・GarageAreaの3つはそれぞれ正の相関がありそう。
* 最終行、または最終列はSalePrice。色が濃いところは相関のあるカラム。OverallQualやGrLivAreaなど。

### SalePriceと相関係数の高い特徴量を見てみる

```python
k = 10
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm,cbar=True,annot=True,square=True,fmt='.2f', annot_kws={'size':10},
                yticklabels=cols.values,
                xticklabels=cols.values)
plt.show()
```

### SalePriceと相関係数の低い特徴量を見てみる

```python
cols = corrmat.nlargest(38, 'SalePrice')['SalePrice'].index[-10:]
cols = cols.append(pd.indexes.base.Index(['SalePrice']))
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm,cbar=True,annot=True,square=True,fmt='.2f', annot_kws={'size':10},
                yticklabels=cols.values,
                xticklabels=cols.values)
plt.show()
```

* SalePriceと最も相関があるのはOverallQual
* GargeCarsとGarageAreaは相関がある
* TotalBsmtSFと1stFlrSFは相関がある

# 単回帰での予測
*  'GrLivArea'を用いて、線形回帰を行う
* GrLivArea: Above grade (ground) living area square feet
* OverallQual: Rates the overall material and finish of the house

```python
from sklearn.linear_model import LinearRegression

X = df[["GrLivArea"]].values
y = df["SalePrice"].values

slr = LinearRegression()
slr.fit(X,y)

print('傾き：{0}'.format(slr.coef_[0]))
print('y切片: {0}'.format(slr.intercept_))
```

図示すると

```python
plt.scatter(X,y)
plt.plot(X,slr.predict(X),color='red')
plt.show()
```
